```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to generate high-res word cloud
def generate_wordcloud(sentiment_label, color):
    text = " ".join(df[df['Sentiment'] == sentiment_label]['Sentence'])
    wc = WordCloud(
        width=1600,         # Higher resolution
        height=800,
        background_color='black',
        colormap=color,
        max_words=200,
        min_font_size=10,
        contour_color='white',
        contour_width=1
    ).generate(text)
    
    plt.figure(figsize=(16, 8))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud - {sentiment_label.capitalize()}", fontsize=20)
    plt.tight_layout(pad=0)
    plt.show()

# Generate word clouds
generate_wordcloud('positive', 'Greens')
generate_wordcloud('negative', 'Reds')
generate_wordcloud('neutral', 'Blues')
```

# Text Preprocessing

```python
#lowercasing text
```

```python
# Create a copy to avoid overwriting original
df['clean_text'] = df['Sentence'].apply(lambda x: x.lower())

# View sample
print(df[['Sentence', 'clean_text']].sample(5))
```

```python
#punctuation cleaning
```

```python
import re

def clean_punctuation(text):
    # Remove all punctuation except $, %, and apostrophes
    return re.sub(r"[^\w\s\$%']", '', text)

df['clean_text'] = df['clean_text'].apply(clean_punctuation)

# View sample
print(df[['Sentence', 'clean_text']].sample(5))
```

```python
#tokenization
```

```python
from nltk.tokenize import word_tokenize

# Tokenize each cleaned sentence
df['tokens'] = df['clean_text'].apply(word_tokenize)

# View sample
print(df[['clean_text', 'tokens']].sample(5))
```

```python
from nltk.corpus import stopwords

# Get standard English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from tokens
df['tokens'] = df['tokens'].apply(lambda tokens: [word for word in tokens if word not in stop_words])

# Preview
print(df[['clean_text', 'tokens']].sample(5))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# Apply lemmatization to each token
df['tokens'] = df['tokens'].apply(lambda tokens: [lemmatizer.lemmatize(word) for word in tokens])

# Optional: Join tokens back into final cleaned sentence
df['final_text'] = df['tokens'].apply(lambda tokens: ' '.join(tokens))

# Preview
print(df[['Sentence', 'final_text']].sample(5))
```

```python
# Select only what you need
df_to_save = df[['final_text', 'Sentiment']]

# Save to CSV
df_to_save.to_csv('/Users/zeeshanabbasi/Desktop/internship projects/financial_sentiment/cleaned_data.csv', index=False)

print("✅ Cleaned dataset saved successfully!")
```

- In this project, we are using FinBERT, a specialized version of BERT pretrained on financial text, to achieve state-of-the-art performance in classifying financial news headlines as positive, negative, or neutral. We selected FinBERT because it understands financial jargon and context far better than traditional methods like TF-IDF or generic language models, making it ideal for capturing the subtle sentiment cues in short news headlines. Instead of manually engineering features or relying on context-free vectorization, FinBERT uses deep contextual embeddings to represent each sentence in a way that captures both the meaning and sentiment nuances. In the next steps, we will load FinBERT using Hugging Face Transformers, tokenize our cleaned headlines using the FinBERT tokenizer, and encode them with attention masks to feed into the model. We’ll then fine-tune FinBERT on our labeled dataset (or alternatively extract embeddings for classification using Logistic Regression or a small neural network). This vectorization process will allow us to turn each headline into a powerful numerical representation that our sentiment classifier can learn from effectively.

```python
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import AdamW, get_scheduler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import numpy as np
```

```python
if torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")

print("Using device:", device)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# FinBERT pretrained model (3-class sentiment)
model_name = "yiyanghkust/finbert-tone"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)

# Send model to MPS device
model.to(device)

print("✅ FinBERT model and tokenizer loaded successfully.")
```

```python
# Load your cleaned data
df = pd.read_csv('/Users/zeeshanabbasi/Desktop/internship projects/financial_sentiment/cleaned_data.csv')

# Encode labels: negative = 0, neutral = 1, positive = 2
label_map = {'negative': 0, 'neutral': 1, 'positive': 2}
df['label'] = df['Sentiment'].map(label_map)

# Confirm mapping
print(df['label'].value_counts())
```

```python
# Split into train and test sets
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['final_text'].tolist(),
    df['label'].tolist(),
    test_size=0.2,
    stratify=df['label'],
    random_state=42
)
```

```python
# Tokenize the text using FinBERT tokenizer
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128, return_tensors='pt')
test_encodings = tokenizer(test_texts, truncation=True, padding=True, max_length=128, return_tensors='pt')
```

```python
from torch.utils.data import Dataset

# Custom dataset class
class FinSentimentDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# Create datasets
train_dataset = FinSentimentDataset(train_encodings, train_labels)
test_dataset = FinSentimentDataset(test_encodings, test_labels)
```

```python
from torch.utils.data import DataLoader

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

print("✅ DataLoaders created.")
```

```python
from transformers import AdamW, get_scheduler

# Define optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)

# Total steps for scheduler
num_training_steps = len(train_loader) * 3  # 3 epochs

# Scheduler for learning rate
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps,
)
```

```python
from tqdm import tqdm

# Put model in training mode
model.train()

# Example loop structure (we'll run this next)
for epoch in range(3):
    loop = tqdm(train_loader, leave=True)
    for batch in loop:
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()

        loop.set_description(f"Epoch {epoch}")
        loop.set_postfix(loss=loss.item())
```

```python
from tqdm import tqdm

# Put model in training mode
model.train()

epochs = 3

for epoch in range(epochs):
    print(f"🔥 Epoch {epoch + 1}/{epochs}")
    loop = tqdm(train_loader, leave=True)
    
    for batch in loop:
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()

        loop.set_description(f"Epoch {epoch + 1}")
        loop.set_postfix(loss=loss.item())
```

```python
model.eval()

preds = []
true_labels = []

with torch.no_grad():
    for batch in test_loader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        logits = outputs.logits
        preds.extend(torch.argmax(logits, axis=1).cpu().numpy())
        true_labels.extend(batch["labels"].cpu().numpy())

# Evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix

print("🔍 Classification Report:\n")
print(classification_report(true_labels, preds, target_names=['negative', 'neutral', 'positive']))

print("\n📊 Confusion Matrix:")
print(confusion_matrix(true_labels, preds))
```
